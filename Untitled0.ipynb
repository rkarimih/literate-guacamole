```python
import numpy as np
from matplotlib import pyplot as plt

ys = np.random.randn(2)
x = np.log(ys)

plt.plot(x, ys,'g')

plt.xlabel('x-value')
plt.ylabel('y-value')

plt.title("model")
plt.show()
```
